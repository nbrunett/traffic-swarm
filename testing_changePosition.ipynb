```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

import vehicle
```

```python
car = vehicle.vehicle(7, 10, 5, 3200, 4, 31.29, 9.8, 12, 12, 0.01, np.pi/3.0, [0, 0], [10, 20])

plotX = list()
plotY = list()
plotSpeed = list()
dt = 0.1
for i in range(100):
    car.changePosition([0, 0], dt)
    plotX.append(car.pos[0])
    plotY.append(car.pos[1])
    plotSpeed.append(car.speed)
    
f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 4))
ax1.plot(range(100), plotX)
ax2.plot(range(100), plotY)
ax3.plot(range(100), plotSpeed)
ax1.set_xlabel('Time')
ax1.set_ylabel('X Position')
ax2.set_ylabel('Y Position')
ax3.set_ylabel('Speed')
```

Looks like the curving in he Speed vs. Time plot comes from the speed in the y-direction being non-zero. If I set it to zero then the speed curve is completely straight. The larger the y-direction speed, relative to the x-direction, the more curved it gets.

```python
car = vehicle.vehicle(7, 10, 5, 3200, 4, 31.29, 9.8, 12, 12, 0.01, np.pi/3.0, [0, 0], [10, 20])

plotX = list()
plotY = list()
plotSpeed = list()
dt = 0.1
for i in range(100):
    car.changePosition([i, 0], dt)
    plotX.append(car.pos[0])
    plotY.append(car.pos[1])
    plotSpeed.append(car.speed)
    
f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 4))
ax1.plot(range(100), plotX)
ax2.plot(range(100), plotY)
ax3.plot(range(100), plotSpeed)
ax1.set_xlabel('Time')
ax1.set_ylabel('X Position')
ax2.set_ylabel('Y Position')
ax3.set_ylabel('Speed')
```

```python
car = vehicle.vehicle(7, 10, 5, 3200, 4, 31.29, 9.8, 12, 12, 0.01, np.pi/2.0, [0, 0], [10, 20])

plotX = list()
plotY = list()
plotSpeed = list()
dt = 0.1
for i in range(100):
    car.changePosition([0, 0], dt)
    plotX.append(car.pos[0])
    plotY.append(car.pos[1])
    plotSpeed.append(car.speed)
    
f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 4))
ax1.scatter(plotX, plotY, color='k')
ax2.scatter(plotX, plotY, color='k')
ax3.scatter(plotX, plotY, color='k')
ax3.quiver(plotX, plotY, np.cos(car.direction), np.sin(car.direction))
ax1.set_xlabel('X Position')
ax1.set_ylabel('Y Position')
ax1.set_xlim([0, 100])
ax1.set_ylim([0, 200])
ax2.set_xlim([0, 50])
ax2.set_ylim([0,100])
ax3.set_xlim([0, 25])
ax3.set_ylim([0, 50])
```
