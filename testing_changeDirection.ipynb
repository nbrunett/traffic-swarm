```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

import vehicle
```

```python
car = vehicle.vehicle(7, 10, 5, 3200, 4, 31.29, 9.8, 12, 12, 0.01, np.pi/2.0, [0, 0], [10, 20])

plotX = list()
plotY = list()
plotDirection = list()
plotSpeed = list()
dt = 0.1
for i in range(200):
    car.changePosition([0, 0], dt)
    if i < 49:
        car.changeDirection((np.pi/2.0)/100.0)
    else:
        car.changeDirection(-(np.pi/2.0)/100.0)
    plotX.append(car.pos[0])
    plotY.append(car.pos[1])
    plotDirection.append(car.direction)
    plotSpeed.append(car.speed)
    
f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 4))
ax1.scatter(plotX, plotY, color='k')
ax2.scatter(plotX, plotY, color='k')
ax2.quiver(plotX, plotY, np.cos(plotDirection), np.sin(plotDirection))
ax3.scatter(plotX, plotY, color='k')
ax3.quiver(plotX, plotY, np.cos(plotDirection), np.sin(plotDirection))
ax1.set_xlabel('X Position')
ax1.set_ylabel('Y Position')
#ax1.set_xlim([0, 100])
#ax1.set_ylim([0, 200])
ax2.set_xlim([-16.5, -16])
ax2.set_ylim([80, 85])
ax3.set_xlim([-20, 20])
ax3.set_ylim([0, 50])
```
